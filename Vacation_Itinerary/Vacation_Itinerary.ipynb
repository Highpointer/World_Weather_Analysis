```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city. DataFrames for each city on the itinerary (10 pt)
# vacation_start = vacation_df.loc["City"] = "Avera"
vacation_start = vacation_df.loc[vacation_df["City"] == "Avera"]
vacation_end = vacation_df.loc[vacation_df["City"] == "Avera"]
vacation_stop1 = vacation_df.loc[vacation_df["City"] == "Eatonton"]
vacation_stop2 = vacation_df.loc[vacation_df["City"] == "Russell"]
vacation_stop3 = vacation_df.loc[vacation_df["City"] == "Marshall"]
vacation_stop4 = vacation_df.loc[vacation_df["City"] == "Muscle Shoals"]
```

temp

```python
# 6. Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function and list indexing.
#start_df = (vacation_start_lat, vacation_start_long).to_numpy()

start = vacation_start["Lat"].to_numpy()[0], vacation_start["Long"].to_numpy()[0]
stop1 = vacation_stop1["Lat"].to_numpy()[0], vacation_stop1["Long"].to_numpy()[0]
stop2 = vacation_stop2["Lat"].to_numpy()[0], vacation_stop2["Long"].to_numpy()[0]
stop3 = vacation_stop3["Lat"].to_numpy()[0], vacation_stop3["Long"].to_numpy()[0]
stop4 = vacation_stop4["Lat"].to_numpy()[0], vacation_stop4["Long"].to_numpy()[0]
end = vacation_end["Lat"].to_numpy()[0], vacation_end["Long"].to_numpy()[0]
print(f"Start: {start}, Stop 1: {stop1}, Stop 2: {stop2}, Stop 3: {stop3}, Stop 4: {stop4}, End: {end}")
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, stop3, and stop4 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING".
# Create Map with route
fig = gmaps.figure()
city_itinerary = gmaps.directions_layer(
        start, end, waypoints=[stop1, stop2, stop3, stop4],
        travel_mode='DRIVING')

fig.add_layer(city_itinerary)
fig #The map below between the cities the travel map is uploaded as WeatherPy_travel_map.png (10 pt)
```

```python
# 8. To create a marker layer map between the five cities.
#  Combine the five city DataFrames into one DataFrame using the concat() function.
# This is the DataFrame that contains the five cities on the itinerary (10 pt)
#itinerary_df = pd.concat([vacation_start, vacation_stop1, vacation_stop2, vacation_stop3, vacation_stop4, vacation_end],ignore_index=True)
itinerary_df = pd.concat([vacation_start, vacation_stop1, vacation_stop2, vacation_stop3, vacation_stop4],ignore_index=True)
itinerary_df
```
